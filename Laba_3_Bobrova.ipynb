```python
import math
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
def kv_function(x):
    return x**2 - 4*x + 2
x=np.linspace(-5,5,400)
y= kv_function(x)
plt.figure(figsize=(5,5))
plt.subplot(2, 1, 1) 
plt.plot(x, y, label=r'$y = x^2 - 4*x + 2$', color='blue')
plt.legend()

plt.subplot(2, 1, 2)
def tr_function(x):
    return 2*np.sin(2*x+1)
x=np.linspace(-5,5,400)
y= tr_function(x)
plt.plot(x, y, label=r'$y = 2*sin(2*x+1)', color='blue')

plt.xlabel('Ось X', fontsize=14)
plt.ylabel('Ось Y', fontsize=14)
plt.legend(loc='upper left', fontsize=8)
plt.tight_layout()
plt.show()
```

```python
def kv_function(x):
    return x**2 - 4*x + 2
def tr_function(x):
    return 2*np.cos(2*x+2)
x1=np.linspace(-5,5,400)
x2=np.linspace(-5,5,400)
y1=kv_function(x1)
y2=tr_function(x2)

plt.figure(figsize=(10,5))
plt.plot(x1, y1, label=r'$y = x^2 - 4*x + 2$', color='blue')
plt.plot(x2, y2, label=r'$y = 2*sin(2*x+1)$', color='red')
plt.xlabel('Ось X', fontsize=14)
plt.ylabel('Ось Y', fontsize=14)
plt.legend(loc='upper left', fontsize=12)
plt.show()
```

```python
t= np.linspace(0, 8 * np.pi, 400)
x= np.cos(3*t)
y=np.sin(3*t)
plt.figure(figsize=(6, 6))
plt.plot(x, y, label=r'$x(t) = \cos(3t), \, y(t) = \sin(3t)$', color='blue')
plt.xlabel('Ось X', fontsize=14)
plt.ylabel('Ось Y', fontsize=14)
plt.legend(loc='upper left', fontsize=10)
plt.show()
```
